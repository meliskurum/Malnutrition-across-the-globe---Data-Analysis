```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Load malnutrition.csv into a pandas DataFrame. 
# The dataset can be found in 'https://www.kaggle.com/ruchi798/malnutrition-across-the-globe/data'
df = pd.read_csv('malnutrition.csv')
df.head()
```

# Data Visualization

```python
income_class = ['upper middle', 'lower middle', 'low', 'high']
levels = [0, 1, 2, 3]
df['Income Classification'].value_counts().plot(kind = 'bar', color = 'g')
plt.xlabel('Income Level', labelpad = 15, size = 15)
plt.ylabel('Number of Countries', labelpad = 15, size = 15)
plt.title('Number of Countries by Income Level', y = 1.02)
plt.xticks(levels, income_class, rotation=0, fontsize=13)
plt.show()
```

```python
# Most of the countries have a upper middle income level. It is followed by lower middle, low and high income levels respectively.
```

### The Relationship Between Income Classification and Severe Wasting

```python
# To visualize the relationship between 'Income Classification' and 'Severe Wasting', scatterplot and boxplot can be used.
```

```python
# 'Income Classification' is an ordinal variable and 'Severe Wasting' is a continuous variable.
```

```python
sns.relplot('Income Classification', 'Severe Wasting', color = 'b', data = df)
```

```python
# The scatterplot above indicates that as the income level gets higher, the percentage of severe wasting(very low weight-for height) decreases.
```

```python
sns.boxplot('Income Classification', 'Severe Wasting', data = df).set_title('The Percentage of Severe Wasting for Different Levels of Income', y =1.1)
```

```python
# According to the boxplot, as the income level gets lower, the variability for the percentage of severe wasting increases.
# All the income levels have high outliers.
```

### The Relationship Between Income Classification and Wasting

```python
sns.relplot('Income Classification', 'Wasting', color = 'r', data = df)
```

```python
# As the income level gets higher the percentage of wasting decreases.
```

```python
sns.boxplot('Income Classification', 'Wasting', data = df).set_title('The Percentage of Wasting for Different Levels of Income',  y=1.1)
```

```python
# Income levels 0(Low income) and 1(Lower Middle Income) have greater variability than 2(Lower Middle Income) and 3(High Income).
# Also, income levels 0 and 2 have high outliers while income levels 1 and 3 do not.
```
